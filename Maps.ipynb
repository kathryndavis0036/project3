```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# URL of the page to scrape
url = "https://www.statista.com/statistics/1380768/sponsorship-revenue-sports-leagues-usa/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table containing the data
    table = soup.find('table')
    
    # Initialize lists to store the data
    leagues = []
    revenues = []

    # Iterate over the rows in the table
    for row in table.find_all('tr')[1:]:  # Skip the header row
        cols = row.find_all('td')
        leagues.append(cols[0].text.strip())
        revenues.append(float(cols[1].text.strip().replace(' billion U.S. dollars', '')))

    # Create a DataFrame
    data = {
        'League': leagues,
        'Revenue (billion U.S. dollars)': revenues
    }
    df = pd.DataFrame(data)

    # Display the DataFrame
    display(df)

    # Save the DataFrame to a CSV file
    df.to_csv('sports_leagues_revenue.csv', index=False)

    # Optionally, save it to an Excel file
    df.to_excel('sports_leagues_revenue.xlsx', index=False)
```

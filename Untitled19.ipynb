```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the webpage containing the tables
url = "https://www.reviews.org/tv-service/live-sports-statistics/"

# Send a GET request to the URL
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Find all tables on the webpage
tables = soup.find_all("table")

# Assuming the second table on the page is the one we want (index 1)
if len(tables) > 1:
    table = tables[1]  # Adjust index if needed
    rows = table.find_all("tr")

    # Initialize CSV writer
    csv_filename = "popular_sports_leagues.csv"
    with open(csv_filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        # Write header row
        header_row = ['#', 'Answer', 'Answer %']
        writer.writerow(header_row)

        # Extract data from each row in the table
        for row in rows[1:]:  # Skip the header row
            cells = row.find_all(["th", "td"])  # Include both <th> and <td> elements
            if len(cells) == 4:  # Ensure we have four columns
                rank = cells[0].text.strip()
                sport = cells[1].text.strip()
                answer = cells[2].text.strip()
                answer_percentage = cells[3].text.strip()

                # Write row to CSV
                writer.writerow([rank, sport, answer, answer_percentage])

    print(f"CSV file '{csv_filename}' has been created successfully.")

else:
    print("No tables found on the webpage.")
```

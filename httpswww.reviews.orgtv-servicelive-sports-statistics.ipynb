```python
import requests
import pandas as pd

url = 'https://www.reviews.org/tv-service/live-sports-statistics/'
html = requests.get(url).content
df_list = pd.read_html(html)
df = df_list[-1]  # Assuming the relevant table is the last one
df.to_csv('my_data.csv')
```

```python
import matplotlib.pyplot as plt

# Example DataFrame (replace this with your actual DataFrame)
df = pd.DataFrame({
    'Sport_id': list(range(1, 21)),
    'Sport_name_&_league': [
        'Football, NFL (National Football League)',
        'Basketball, NBA (National Basketball Association)',
        'Baseball, MLB (Major League Baseball)',
        'Football, NCAA College Football',
        'Basketball, NCAA College Basketball (Men\'s and Women\'s)',
        'Auto racing, NASCAR (National Association for Stock Car Auto Racing)',
        'Hockey, NHL (National Hockey League)',
        'Auto racing, F1 (Formula 1)',
        'Combat Sports, UFC (Ultimate Fighting Championship)',
        'Combat Sports, Bellator MMA (Bellator Mixed Martial Arts)',
        'Basketball, WNBA (Women\'s National Basketball Association)',
        'Soccer, MLS (Major League Soccer)',
        'Basketball, NBA G League (NBA Development League)',
        'Auto racing, IndyCar (IndyCar Series)',
        'Combat Sports, WWE (World Wrestling Entertainment)',
        'Golf, PGA (Professional Golfers\' Association)',
        'Basketball, WNBA (Women\'s National Basketball Association)',
        'Combat Sports, AEW (All Elite Wrestling)',
        'Combat Sports, WBA (World Boxing Association)',
        'Auto racing, NCAA College Baseball'
    ],
    'Sports_percentage': [65, 51, 42, 30, 25, 24, 20, 17, 17, 13, 12, 12, 12, 11, 11, 10, 9, 9, 8, 8]
})

# Plotting the bar chart
plt.figure(figsize=(12, 8))
bars = plt.barh(df['Sport_name_&_league'], df['Sports_percentage'], color='skyblue')

# Adding labels
plt.xlabel('Sports Percentage (%)')
plt.ylabel('Sport Name & League')
plt.title('Sports Percentage Distribution')

# Invert y-axis to display highest percentage at the top
plt.gca().invert_yaxis()

# Adding values on top of bars
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width()}%', 
             va='center', ha='left', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```
